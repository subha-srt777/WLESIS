```r
# ── Install packages (run once only, ever) ──────
install.packages("tidyverse")  # data wrangling
install.packages("psych")      # descriptive stats

# ── Load packages (run every session) ───────────
library(tidyverse)
library(psych)

# ── Load your CSV file ──────────────────────────
# Put your CSV in the same folder as your R script
data <- read_csv("dhaka_bus_100trips.csv")

# ── First look: structure ────────────────────────
glimpse(data)
# Shows: how many rows, columns, what type each is

# ── Count your sample ───────────────────────────
data %>% summarise(
  n_drivers    = n_distinct(driver_id),
  n_passengers = n()
)

# ── Calculate composite scores ──────────────────
# Average the 16 WLEIS items into subscales + total EI
# Average the 15 pe items into one experience score
data <- data %>%
  mutate(
    SEA = rowMeans(select(., wleis1:wleis4)),
    OEA = rowMeans(select(., wleis5:wleis8)),
    UOE = rowMeans(select(., wleis9:wleis12)),
    ROE = rowMeans(select(., wleis13:wleis16)),
    EI  = rowMeans(select(., wleis1:wleis16)),
    passenger_experience = rowMeans(
                             select(., pe1:pe15))
  )

# ── Descriptive statistics ───────────────────────
# The main table you'll put in your thesis
describe(data %>% select(
  EI, SEA, OEA, UOE, ROE,
  noise_db, jerking_index, crowd_density,
  traffic_index, ttc_mean,
  passenger_experience, facial_valence,
  conductor_rating, age
))
```

```r
install.packages("lavaan")
library(lavaan)

# ── Get one row per driver ─────────────────────
driver_data <- data %>%
  distinct(driver_id, .keep_all = TRUE)

# ── Define the model ──────────────────────────
# This is where you tell R the 4-factor structure
# =~ means "is measured by"

wleis_model <- '
  SEA =~ wleis1 + wleis2 + wleis3 + wleis4
  OEA =~ wleis5 + wleis6 + wleis7 + wleis8
  UOE =~ wleis9 + wleis10 + wleis11 + wleis12
  ROE =~ wleis13 + wleis14 + wleis15 + wleis16
'

# ── Fit the model ─────────────────────────────
# MLR = robust estimator (better for Likert data)
cfa_fit <- cfa(
  wleis_model,
  data      = driver_data,
  estimator = "MLR"
)

# ── See results ───────────────────────────────
summary(cfa_fit,
  fit.measures  = TRUE,   # shows CFI, RMSEA, SRMR
  standardized  = TRUE    # shows factor loadings
)

# ── If fit is bad, check what to fix ──────────
modindices(cfa_fit, sort = TRUE, maximum.number = 10)
# This suggests which items might belong to
# a different factor in your population
```

```r
install.packages("writexl")
library(writexl)

# Capture the output of summary(cfa_fit)
summary_output <- capture.output(summary(cfa_fit,
  fit.measures  = TRUE,
  standardized  = TRUE
))
cfa_results_list <- list()
cfa_results_list[["CFA_Summary"]] <- data.frame(Output = summary_output)

# Capture the output of modindices(cfa_fit)
modindices_output <- capture.output(modindices(cfa_fit, sort = TRUE, maximum.number = 10))
cfa_results_list[["Modification_Indices"]] <- data.frame(Output = modindices_output)

# Save the results to an Excel file
write_xlsx(cfa_results_list, "cfa_results.xlsx")
```

```r
install.packages("brms")
install.packages("tidybayes")
install.packages("bayesplot")
library(brms); library(tidybayes); library(bayesplot)

# ── Step 1: Define priors ─────────────────────
# Tell R what we already know from EI literature
priors <- c(
  # EI: expect positive effect ~0.28 (from meta-analysis)
  prior(normal(0.28, 0.10), class = "b", coef = "EI"),
  # Other predictors: unsure of magnitude, use weak prior
  prior(normal(0, 0.5),    class = "b", coef = "noise_db"),
  prior(normal(0, 0.5),    class = "b", coef = "jerking_index"),
  prior(normal(0, 0.5),    class = "b", coef = "crowd_density"),
  prior(normal(0, 0.5),    class = "b", coef = "traffic_index"),
  prior(normal(0, 0.5),    class = "b", coef = "ttc_mean"),
  # Variance components: must be positive
  prior(normal(0, 1),     class = "sd"),
  prior(normal(0, 1),     class = "sigma")
)

# ── Step 2: Fit the model ─────────────────────
# passenger_experience = outcome
# (1 | driver_id) = random intercept per driver
bayesian_model <- brm(
  passenger_experience ~
    EI + noise_db + jerking_index +
    crowd_density + traffic_index + ttc_mean +
    (1 | driver_id),
  data    = data,
  prior   = priors,
  family  = gaussian(),
  chains  = 4,     # 4 independent calculation chains
  iter    = 4000,  # 4000 iterations per chain
  warmup  = 2000,  # first 2000 discarded (warm-up)
  cores   = 4,     # use 4 CPU cores (faster)
  seed    = 42     # for reproducibility
)

# ── Step 3: See results ───────────────────────
summary(bayesian_model)

# ── Step 4: What's the probability EI is positive?
posterior <- as_draws_df(bayesian_model)
mean(posterior$b_EI > 0)
# e.g. 0.94 → "94% probability EI has positive effect"
```

```r
install.packages("writexl")
library(writexl)

# Capture the output of summary(bayesian_model)
summary_bayesian_output <- capture.output(summary(bayesian_model))
bayesian_results_list <- list()
bayesian_results_list[["Bayesian_Model_Summary"]] <- data.frame(Output = summary_bayesian_output)

# Capture the output of mean(posterior$b_EI > 0)
posterior_prob_output <- capture.output(mean(posterior$b_EI > 0))
bayesian_results_list[["EI_Positive_Probability"]] <- data.frame(Output = posterior_prob_output)

# Save the results to an Excel file
write_xlsx(bayesian_results_list, "bayesian_model_results.xlsx")
```

```r
install.packages("irr")
library(irr)

# ── Step 1: Create discordance variable ───────
# "above midpoint" = positive direction
# Midpoint of 1-5 scale = 3
# Facial valence: positive > 0, negative < 0

data <- data %>%
  mutate(
    # Direction of self-report: 1=positive, 0=negative
    sr_direction  = ifelse(self_report_comfort > 3,
                           "positive", "negative"),
    # Direction of face: 1=positive, 0=negative
    fv_direction  = ifelse(facial_valence > 0,
                           "positive", "negative"),
    # Discordant = directions don't match
    discordant    = ifelse(
      sr_direction != fv_direction, 1, 0
    )
  )

# ── Step 2: Cohen's Kappa ─────────────────────
# Overall agreement between self-report and face
kappa2(data[, c("sr_direction", "fv_direction")])

# ── Step 3: Who is discordant? ────────────────
# Count discordance by gender
data %>%
  group_by(gender) %>%
  summarise(
    n            = n(),
    n_discordant = sum(discordant),
    pct          = mean(discordant) * 100
  )

# ── Step 4: Bayesian logistic regression ──────
# Does gender predict probability of discordance?
discordance_model <- brm(
  discordant ~ gender + EI + (1 | driver_id),
  data   = data,
  family = bernoulli(link = "logit"),
  # binary outcome (0 or 1) needs bernoulli family
  chains = 4, iter = 4000,
  warmup = 2000, cores = 4, seed = 42
)

summary(discordance_model)

# ── Step 5: Probability female are more discordant
disc_post <- as_draws_df(discordance_model)
mean(disc_post$b_genderFemale > 0)
# e.g. 0.91 → females 91% more likely to be discordant
```
